```python
import pandas as pd
# Set the option to display all columns
pd.set_option('display.max_columns', None)
```

```python
for year in range(2005, 2025):
    crashes = pd.read_csv(f"/home/tonydeals/Downloads/penn/CRASH_{year}.csv")
    vehicles = pd.read_csv(f"/home/tonydeals/Downloads/penn/VEHICLE_{year}.csv")
    persons = pd.read_csv(f"/home/tonydeals/Downloads/penn/PERSON_{year}.csv")
    roadways = pd.read_csv(f"/home/tonydeals/Downloads/penn/ROADWAY_{year}.csv")
    severe_crash_ids = persons[(persons["INJ_SEVERITY"]==1) | (persons["INJ_SEVERITY"]==2)]['CRN']
    crashes = crashes[crashes['CRN'].isin(severe_crash_ids)]
    crashes = crashes[['CRN', 'DEC_LATITUDE', 'DEC_LONGITUDE', 'COUNTY', 'MUNICIPALITY', 'CRASH_MONTH', 'DAY_OF_WEEK', 'ARRIVAL_TM','TIME_OF_DAY', 'SUSP_SERIOUS_INJ_COUNT', 'FATAL_COUNT', 'COLLISION_TYPE']]
    vehicles = vehicles[vehicles['CRN'].isin(severe_crash_ids)]
    vehicles = vehicles[['CRN', 'UNIT_NUM', 'BODY_TYPE', 'VEH_TYPE', 'UNIT_TYPE', 'MODEL_YR', 'MAKE_CD', 'NON_MOTORIST', 'VINA_BODY_TYPE_CD', 'DVR_PRES_IND']]
    persons = persons[persons['CRN'].isin(severe_crash_ids)]
    persons = persons[['CRN', 'AGE', 'SEX', 'INJ_SEVERITY', 'UNIT_NUM', 'PERSON_TYPE']]
    roadways = roadways[roadways['CRN'].isin(severe_crash_ids)]
    roadways = roadways[['CRN', "STREET_NAME"]]
    crashes.to_csv(f"/home/tonydeals/app/ntsb/pennsylvania/crash_{year}.csv")
    vehicles.to_csv(f"/home/tonydeals/app/ntsb/pennsylvania/vehicle_{year}.csv")
    persons.to_csv(f"/home/tonydeals/app/ntsb/pennsylvania/person_{year}.csv")
    roadways.to_csv(f"/home/tonydeals/app/ntsb/pennsylvania/roadway_{year}.csv")
    # break
    
```
