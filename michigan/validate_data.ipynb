```python
import pandas as pd
import os
```

```python
start = 1
end = 90
total = int(os.listdir("data/2004")[0].split("of_")[1].split(".")[0])
all_person_ids = set()
```

```python
for year in range(2004,2025):
    start = 1
    end = 90
    total = int(os.listdir(f"data/{year}")[0].split("of_")[1].split(".")[0])
    all_person_ids = set()
    
    while start < total:
        if end > total:
            end = total    
        table = pd.read_csv(f"data/{year}/crash_data_{year}__{start}_to_{end}_of_{total}.csv", skiprows=1)
        for x in table.index:
            new_id = int(table['Person Instance'][x])
            if new_id in all_person_ids:
                print("DATA ERROR DATA ERROR DATA ERROR")
                print(year)
                print(f"{start} to {end} of {total}")
                break
            all_person_ids.add(new_id)
        start += 90
        end += 90
```

```python

with open("../data/fatalities/management/commands/michigan_2004.py", 'r') as src:
    content = src.read()
# 2. Swap the strings using the replace() method
# Note: .replace() returns a new string and does not modify the original
for year in range(2005,2025):
    modified_content = content.replace("2004", str(year))
# 3. Write the modified content to the new duplicate file
    with open(f"../data/fatalities/management/commands/michigan_{year}.py", 'w') as dst:
        dst.write(modified_content)
```
