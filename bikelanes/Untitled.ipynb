```python
import geopandas as gpd
from shapely import wkt
import pandas as pd
```

```python
bikelanes = gpd.read_file("bike.csv")
bikelanes['the_geom'] = bikelanes['the_geom'].apply(wkt.loads)
bikelanes = bikelanes.set_geometry("the_geom")
protected = bikelanes[(bikelanes['Bike Level of Comfort']=="HU") | (bikelanes['Bike Level of Comfort']=="H") | (bikelanes['Bike Level of Comfort']=="HP")| (bikelanes['Bike Level of Comfort']=="SS")]
protected.to_file("bikelanes.geojson", driver="GeoJSON")
```

```python
carlanes = gpd.read_file("car.csv")
carlanes['the_geom'] = carlanes['the_geom'].apply(wkt.loads)
carlanes = carlanes.set_geometry("the_geom")
carlanes.to_file("carlanes.geojson", driver="GeoJSON")
```
