```python
import pandas as pd
import zipfile
from datetime import datetime
pd.set_option('display.max_columns', None)
```

```python
for year in range(2006,2017):
    # EXTRACT DATA FROM ZIPFILES
    extract_path = f"{year}"
    zip_path = f"NewJersey{year}Accidents.zip"
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(extract_path)
    zip_path = f"NewJersey{year}Vehicles.zip"
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(extract_path)
    zip_path = f"NewJersey{year}Drivers.zip"
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(extract_path)
    zip_path = f"NewJersey{year}Occupants.zip"
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(extract_path)
    zip_path = f"NewJersey{year}Pedestrians.zip"
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(extract_path)
    # Comb person-level CSVs to get all injury crash IDs
    injuries = set()
    deaths = set()
    with open(f"{year}/NewJersey{year}Pedestrians.txt", 'r') as file:
        for line in file:
            condition = line[35:37].strip()
            crash_id = line[:31].strip()
            if condition and int(condition) == 2:
                injuries.add(crash_id)
            if condition and int(condition) == 1:
                deaths.add(crash_id)
    
    with open(f"{year}/NewJersey{year}Occupants.txt", 'r') as file:
        for line in file:
            condition = line[38:40].strip()
            crash_id = line[:31].strip()
            if condition and int(condition) == 2:
                injuries.add(crash_id)
            if condition and int(condition) == 1:
                deaths.add(crash_id)
    # Create a set of crashes that contain a serious injury but no deaths
    injurycrashes = injuries - deaths

    with open(f"{year}/crash{year}.csv", "w") as f:
        f.write("id,dt,lat,lon,city,county,street_1,street_2,crash_type,death_count,severity\n")
        with open(f'{year}/NewJersey{year}Accidents.txt', 'r', encoding='latin-1') as file:
            for line in file:
                coords = line[353:370]
                crash_id = line[:31].strip()
                if coords != "       ,        ," and crash_id in injurycrashes:
                    total_killed = int(line[134:136])
                    total_injured = int(line[137:139])
                    severity = line[146:147]
                    date = line[70:80]
                    time = line[84:88].strip()
                    if not time:
                        time = "0000"
                    dt = datetime.strptime(date + " " + time, "%m/%d/%Y %H%M")
                    latitude = line[352:360]
                    longitude = line[361:369]
                    county = line[32:44].strip()
                    city = line[45:69].strip()
                    street_1 = line[160:210].strip()
                    street_2 = line[279:314].strip()
                    crash_type = line[154:156].strip()
                    f.write(f"{crash_id},{dt},{latitude},{longitude},{city},{county},{street_1},{street_2},{crash_type},{total_killed},{severity}\n")
    
    with open(f"{year}/pedestrian{year}.csv", "w") as f:
        f.write("id,bike,age,sex,condition\n")
        with open(f'{year}/NewJersey{year}Pedestrians.txt', 'r', encoding='latin-1') as file:
            for line in file:
                crash_id = line[:31].strip()
                if crash_id in injurycrashes:
                    condition = line[35:37]
                    age = line[84:87].strip()
                    sex = line[88:89]
                    bike = line[197:198]
                    f.write(f"{crash_id},{bike},{age},{sex},{condition}\n")
    
    with open(f"{year}/occupant{year}.csv", "w") as f:
        f.write("id,vehicle_number,position,occupant_number,age,sex,condition\n")
        with open(f'{year}/NewJersey{year}Occupants.txt', 'r', encoding='latin-1') as file:
            for line in file:
                crash_id = line[:31].strip()
                if crash_id in injurycrashes:
                    condition = line[38:40]
                    age = line[47:50].strip()
                    sex = line[51:52]
                    vehicle_number = line[32:34].strip()
                    occupant_number = line[35:37].strip()
                    position = line[41:43]
                    f.write(f"{crash_id},{vehicle_number},{position},{occupant_number},{age},{sex},{condition}\n")
    
    with open(f"{year}/vehicle{year}.csv", "w") as f:
        f.write("id,vehicle_number,year,make,model,hit_run\n")
        with open(f'{year}/NewJersey{year}Vehicles.txt', 'r', encoding='latin-1') as file:
            for line in file:
                crash_id = line[:31].strip()
                if crash_id in injurycrashes:
                    vehicle_number = line[32:34].strip()
                    make = line[43:73].strip()
                    model = line[74:94].replace(",","").strip()
                    yr = line[99:103]
                    hit_run = line[239:240]
                    f.write(f"{crash_id},{vehicle_number},{yr},{make},{model},{hit_run}\n")
    
    with open(f"{year}/driver{year}.csv", "w") as f:
        f.write("id,vehicle_number,charge\n")
        with open(f'{year}/NewJersey{year}Drivers.txt', 'r', encoding='latin-1') as file:
            for line in file:
                crash_id = line[:31].strip()
                if crash_id in injurycrashes:
                    vehicle_number = line[32:34].strip()
                    violation = line[95:125]
                    f.write(f"{crash_id},{vehicle_number},{violation}\n")
```
